```python
# Converting data to numpy array
wine_array =  wine.values
wine_array
```

```python
# Normalizing the numerical data
wine_norm = scale(wine_array)
wine_norm
```

```python
# Applying PCA fit transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
# PCA Components matrix or convariance matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
#Cumalative varaince of each PCA
var = np.cumsum(np.round(var,decimals=4)*100)
var
```

```python
plt.plot(var,color="blue")
```

```python
# Final dataframe
final_df=pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Checking with other Clustering algorithms
#Hierarchical Clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create dendrograms
plt.figure(figsize=(10,8))
dendrograms=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# create clusters
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
wine1=wine.copy()
wine1['clustersid']=hclusters.labels_
wine1
```

```python
#K -Means Clustering
from sklearn.cluster import KMeans
```

```python
#As we already have normalized data
#use elbow curve graph to find optimum no of clusters (k value)from k values range
#k-means algo aims to choose centroids that minimize the inertia or within cluster sum-of-
#random state can be anything from 0 to 42, but the same number to be used everytime,so that the reuse
```

```python
#within-cluster sum-of-squares criterion
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
    
```

```python
#Plot K values range vs WCSS(within-cluster-sum-of -sqaures) to get elbow graph for choosing K(no.of.clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
# Build cluster using k=3
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
wine3=wine.copy()
wine3['clusters3id']=clusters3.labels_
wine3
```
